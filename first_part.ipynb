```python
#determine the room and defining the variables
r0 = 2.5 #growth constant w/ no social distancing, no masks
r1 = 1.25 #growth constant w/ 50% less exposure, no masks
r2 = .625 #growth constant w/ 75% less exposure, no masks 
room = input("Pick a building to estimate time until 100% positive cases: UU202 (34) or LH01 (85) or GW69EX (66)")
if room == 'UU202':
    capacity = 34 #set the max capacity in the room
elif room == 'LH01':
    capacity = 85
elif user1 == 'GW69EX':
    capacity = 66
else:
    print('Incorrect input. Please enter in a correct room.')
```

```python
#calculate how long it will take if only one person is infected but with no masks
from math import *
import matplotlib.pyplot as plt

t1 = log(capacity)/r0 #t measured in minutes, no social distancing, no mask
t2 = log(capacity)/r1 #50% social distancing, no mask
t3 = log(capacity)/r2 #75% social distancing, no mask

labels = ['t1', 't2', 't3']
sizes = [t1, t2, t3]
colors = ['red', 'green' , 'blue']
explode = (0, 0.1, 0)  # Only "explode" the 2nd slice
fig = plt.figure(figsize=(5, 4), dpi=80)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%0.1f%%', shadow=True, startangle=0)
plt.title("Infection Rates with No Masks", bbox={'facecolor':'0.8', 'pad':1})
plt.axis('equal')  # Set aspect ratio to be equal so the pie is a circle.
plt.show()
```

```python
#calculate how long it will take if only one person is infected with masks
#From source: Cloth masks provide 67% protection. Factor in the 33% risk.
import pandas as pd
r3 = .33*r0
r4 = .33*r1
r5 = .33*r2

t4 = log(capacity)/r3 #no social distancing, with mask
t5 = log(capacity)/r4 #50% social distancing, with mask
t6 = log(capacity)/r5 #75% social distancing, with mask

labels = ['t4', 't5', 't6']
sizes = [t4, t5, t6]
colors = ['pink', 'blue' , 'purple']
explode = (0, 0, 0.1)  # Only "explode" the 3rd slice
fig = plt.figure(figsize=(5, 4), dpi=80)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%0.1f%%', shadow=True, startangle=0)
plt.title("Infection Rates with Masks", bbox={'facecolor':'0.8', 'pad':1})
plt.axis('equal')  # Set aspect ratio to be equal so the pie is a circle.
plt.show()

t1_2 = format(t1, '.2f') #rounds each of the minutes to the 2nd decimal place 
t2_2 = format(t2, '.2f')
t3_2 = format(t3, '.2f')
t4_2 = format(t4, '.2f')
t5_2 = format(t5, '.2f')
t6_2 = format(t6, '.2f')

#creates dataframe for no masks and masks at 0%, 50%, and 75% social distancing
frame1 = {'0%': pd.Series([t1_2, t4_2], index=['No Masks', 'With Masks']),
          '50%': pd.Series([t2_2, t5_2], index=['No Masks', 'With Masks']),
          '75%': pd.Series([t3_2, t6_2], index=['No Masks', 'With Masks'])}
df1 = pd.DataFrame(frame1)
df1
```
